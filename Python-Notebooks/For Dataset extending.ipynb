```python
import pylab
import calendar
import numpy as np
import pandas as pd

from datetime import datetime
import matplotlib.pyplot as plt
import warnings
pd.options.mode.chained_assignment = None
warnings.filterwarnings("ignore", category=DeprecationWarning)
%matplotlib inline
```

```python
df = pd.read_csv("2010-q4_trip_history_data.csv")
```

```python
df.shape
```

```python
df.dtypes
df = df.assign(instant=[ 1+ i for i in range(len(df))])[['instant'] + df.columns.tolist()]
df
```

```python
def f(row):
   
    if row["date"].split("-")[1] in ["01","02","03"]:
        season = 1
    elif row["date"].split("-")[1] in ["04","05","06"]:
        season = 2
    elif row["date"].split("-")[1] in ["07","08","09"]:
        season = 3
    else:
        season = 4

    if (row["date"].split("-")[1] == "03") and (int(row["date"].split("-")[2]) > 19):
        season = 2
    elif (row["date"].split("-")[1] == "06") and (int(row["date"].split("-")[2]) > 20):
        season = 3
    elif (row["date"].split("-")[1] == "09") and (int(row["date"].split("-")[2]) > 21):
        season = 4
    elif (row["date"].split("-")[1] == "12") and (int(row["date"].split("-")[2]) > 20):
        season = 1

    return season
```

```python
def h(row):
    val = row["date"].split("-")[2] + row["date"].split("-")[1]
    if val in ["0101","1501","1902","1604","2805","0407","0309","0810","1211","2512","3112"]:
       return 1
    else:
        return 0
        
```

```python
def g(row):
    val = df[df["date"]==row["date"]& df["Member type"]=="Casual"].count()
    return val
    
```

```python
def yr(row):
    val = int(row["date"].split("-")[0])-2010
    return val    
```

```python
def mn(row):
    val = int(row["date"].split("-")[1])
    return val    
```

```python
def week(row):
    val = calendar.weekday(int(row["date"].split("-")[0]),int(row["date"].split("-")[1]),int(row["date"].split("-")[2]))
    return val
```

```python
df["date"] = df.Startdate.apply(lambda x : x.split()[0])
df["hour"] = df.Startdate.apply(lambda x : x.split()[1].split(":")[0])
df["weekday"] = df.apply(week,axis=1)
df["month"] = df.date.apply(lambda dateString : calendar.month_name[datetime.strptime(dateString,"%Y-%m-%d").month])
df["holiday"] = df.apply(h,axis =1)
df["season"] = df.apply(f,axis =1)
df["yr"] = df.apply(yr,axis =1)
df["mnth"] = df.apply(mn,axis =1)
df.drop(['Startdate','Duration (ms)', 'End date','Start station number','Start station','End station number','End station','Bike number'], axis=1, inplace=True)
df.groupby(['date','Member type'])['Member type'].count()
```
