```python
with open('dataset/val.txt','r') as f:
    for line in f:
        words+=line.split()
```

```python
wordfreq = {}
for raw_word in words:
    if raw_word not in wordfreq:
        wordfreq[raw_word] = 0 
    wordfreq[raw_word] += 1
```
