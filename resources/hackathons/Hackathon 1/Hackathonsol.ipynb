### Function to create dictionary

* Convert uppercase to lower
* Replace `,|.|!|?|'|'s|)|(|=|;` with `''`

```python
# def createDict(self):
f=open("dataset/val.txt","r")
fn=f.read()
fn=fn.lower()
fn= re.sub(r"[?./-=!]", ' ',fn)
dic={}

# IMplement regex

fn=fn.split()
for i in range(len(fn)):
    if fn[i] in dic.keys():
        dic[fn[i]]=dic[fn[i]]+1
    else:
        dic[fn[i]]=1
print (dic)
            

```

### Counting occurences of words

```python
# def countOccure(self,d):
values=dic.values()
valuesum=sum(values)
listperc=[]
c=[(v, k) for k, v in dic.items()]
c.sort()
c.reverse()
c=[(k, v) for v, k in c]
for i in range(10):
    
    listperc.append(c[i][1]/valuesum)
    print(c[i])
print (listperc)

    
```

```python
fields=['Index','Word','Occurences',"PercentOccur"]
l=[]
for i in range(10):
    o=[]
    o.append(i+1)
    o.append(c[i][0])
    o.append(c[i][1])
    o.append(listperc[i])
    print(o)
    l.append(o)
filename="record.csv"
with open(filename, 'w') as csvfile:
    csvwriter = csv.writer(csvfile)         
    csvwriter.writerow(fields)     
    csvwriter.writerows(l)
```
