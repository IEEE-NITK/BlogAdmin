# Seaborn


Seaborn is a library for making attractive and informative statistical graphics in Python. It is built on top of matplotlib. It provides a high-level interface to draw statistical graphics.

To get started with seaborn, youâ€™re going to need to install it in the terminal with either ** pip install seaborn ** or ** conda install seaborn **. Then simply include import seaborn as sns at the top of your python file.

### Special features

1)Several built-in themes for styling matplotlib graphics 

2)Tools for choosing color palettes to make beautiful plots that reveal patterns in your data

3)Functions for visualizing univariate and bivariate distributions or for comparing them between subsets of data
    

### Loading data to construct seaborn plots



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris data
iris = sns.load_dataset("iris")

# Construct iris plot
sns.barplot(x="species", y="petal_length", data=iris)

# Show plot
plt.show()
```

### Visualizing pairwise relationships in a dataset

To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame. by default, it also draws the univariate distribution of each variable on the diagonal Axes:

```python
sns.pairplot(iris)
plt.show()
```

### Seaborn themes

###### The default theme is darkgrid. Here we change the theme to whitegrid

```python
# Set theme
sns.set_style('whitegrid')
 
# Violin plot
sns.boxplot(x='species', y='petal_length', data=iris)

plt.show()
```

### Colour palette for Seaborn

Seaborn has some nice built-in color palette features. It also allows us to set custom color palettes. We can simply create an ordered Python list of color hex values.

```python
# global config settings to control things like style, font size, color palette etc.
sns.set(context="notebook", style="darkgrid", palette="dark")
# seaborn has some nice built-in color palette features
sns.palplot(sns.color_palette())
plt.show()
```

```python

# Custom palettes
type_colors = ['#78C850',  
               '#F08030',  
                '#6890F0', 
                '#A8B820',  
                '#A8A878',  
                '#A040A0', 
                '#F8D030', 
                    
                   ]
sns.boxplot(x='species', y='petal_length', data=iris, 
               palette=type_colors) # Set color palette
plt.show()
```

### What all can seaborn do ?

#### FacetGrid
It shows bivariate scatterplots and also specifies which species each plant belongs to

```python
sns.FacetGrid(iris, hue="species",size=6) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend()

plt.show()
```

#### Boxplot

We can look at an individual feature in Seaborn through a boxplot.

```python
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.show()
```
